# Pandas

Pandas is a Python library. A library in Python contains a set of tools (called functions) that perform tasks on our data.

Pandas is one of the best options for working with tabular data in Python.


# Dataset Overview

Dataset from the BCO-DMO repository that we are working with today: https://www.bco-dmo.org/dataset/967649

Summary: 
This dataset contains CTD and in situ sonde water-column measurements from Lake Superior and Lake Erie, collected during 22 cruises from May 2021 to May 2024 at five stations.
The data include depth-profile measurements collected with a SeaBird 911plus CTD profiler and a Eureka Manta 2 Sonde; the main file is a processed CSV, with raw CTD files also provided as supplemental data.
CTD casts were lowered from research vessels through the water column, while sonde measurements were recorded every meter during downward profiles from the side of the vessel.

# Import a library

Python doesn’t load all of the libraries available to it by default. We have to add an import statement to our code in order to use library functions. To import a library, we use the syntax import libraryName. If we want to give the library a nickname to shorten the command, we can add as nickNameHere. The common nickname for pandas is pd.

```python
#Import the library
import pandas as pd
```

## Importing data & the Pandas Dataframe

Pandas provides an object called DataFrame. Dataframes represent tabular data. They are a 2-dimensional data structure. They include columns, each of which is a Series with a name, and all columns share the same index (An index refers to the position of an element in the data structure.). We import a csv table into a data frame using the read_csv function.

```python
# Create a dataframe by reading a CSV from a BCO-DMO URL
data_url_BCODMO = 'https://datadocs.bco-dmo.org/dataset/967649/file/nGqVpmBU6pYynm/967649_v1_ctd_sonde_great_lakes.csv'

my_dataframe = pd.read_csv(data_url_BCODMO)

# Look at the first few rows
my_dataframe.head(5)
```

## Plotting specific columns
Plot temperature (y-axis) over depth (x-axis)

```python
# import plotting library matplotlib
import matplotlib.pyplot as plt
```

```python
# Plot temperature over depth

# Create a new figure.
# figsize=(6, 8) means the figure will be 6 inches wide and 8 inches tall.
plt.figure(figsize=(6, 8))

# Make the plot.
# The first value, df["temperature"], goes on the x-axis.
# The second value, df["depth"], goes on the y-axis.
#
# This means:
# x-axis = temperature
# y-axis = depth
plt.plot(my_dataframe["Temperature"], my_dataframe["Depth"])

# Add a label to the x-axis.
plt.xlabel("Temperature")

# Add a label to the y-axis.
plt.ylabel("Depth")

# Add a title to the plot.
plt.title("Temperature over Depth")

# In oceanography and limnology, depth is usually shown increasing downward.
# By default, matplotlib puts smaller y-values at the bottom and larger y-values at the top.
# This command flips the y-axis so 0 depth is near the top
# and deeper measurements appear lower on the plot.
plt.gca().invert_yaxis()

# Add grid lines to make the plot easier to read.
plt.grid(True)

# Display the plot.
plt.show()
```
